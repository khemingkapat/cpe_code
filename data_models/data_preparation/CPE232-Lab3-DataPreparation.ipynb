[Q10] Complete the function to apply `ratio` column that calculates from `First Class` and `Standard Class` columns on `pivot_table_result`


```python
# [Q10] Complete the function to apply `ratio` column that calculates from `First Class` and `Standard Class` columns on `pivot_table_result`

# function to transform the ratio
def get_class_ratio(row):
    
    # get the first class column
    first_class = row['First Class']
    
    # get the standard class column
    standard_class = row['Standard Class']
    
    # calculate the ratio
    ratio = first_class / standard_class
    
    return ratio

pivot_table_result["ratio"] = pivot_table_result.apply(get_class_ratio, axis=1)
    
pivot_table_result.head()
```

[Q11] After complete Q10, What does the `apply` function do?
applying the function by find the first class count / standard class count along every row

[Q12] Create a new column(`short_ratio`) that works the same as Q10 but with `lambda` function


```python
# [Q12] Create a new column(`short_ratio`) that works the same as Q10 but with `lambda` function

pivot_table_result["short_ratio"] = pivot_table_result.apply(lambda row: row['First Class']/row['Standard Class'], axis=1)

pivot_table_result.head()
```

[Q13] What is the difference between `apply` and `lambda` ? give 2 examples use case.
lambda is the anonymous function that create function without name, `apply` is `pandas` method of dataFrame that require any kind of function (whether `def` or `lambda` function to apply to all, when specify axis)

simple use case as applying log summation of value in each row as `df.apply(lambda row : row.A + row.B,axis = 1)` as summing col `A` and `B` in each row together or some column as `df.apply(np.sum,axis=0)` as summing row in each column together
