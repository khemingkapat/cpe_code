## Task 2

Given a dictionary of students with their scores, find the average score of each student across all subjects and identify the student with the highest average score.

```python
students_grades = {
    "John": [85, 90, 78],
    "Alice": [88, 92, 80],
    "Bob": [75, 85, 72],
    "Diana": [90, 95, 94],
    "Charlie": [70, 65, 80]
}
```

```python
# Calculate and print the average grade for each student
students_gpa = dict()
for name,grades in students_grades.items():
    gpa = sum(grades)/len(grades)
    students_gpa[name] = gpa
    print(name,gpa)
```

```python
# Find the student with the highest average grade
print(max(students_gpa.items(),key=lambda p : p[1]))
```

## Average `MonthlyRate` Different 

```python
from scipy import stats
import numpy as np
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Define criteria
criterias = ["Gender", "MaritalStatus", "EducationField"]

# Create a 3-row, 2-column subplot layout
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(8, 8))

# Iterate over criteria and plot bar and box plots
for i, criteria in enumerate(criterias):
    # Bar plot (left column)
    df.groupby([criteria])["MonthlyRate"].mean().plot(kind="bar", ax=axes[i, 0])
    axes[i, 0].set_title(f"Average MonthlyRate by {criteria}")
    axes[i, 0].set_ylabel("Average Monthly Rate")
    axes[i,0].tick_params(axis='x', labelrotation=45)

    for criteria_item in df[criteria].unique():
        data = df[df[criteria] == criteria_item].MonthlyRate
        mu, sigma = stats.norm.fit(data)
        x = np.linspace(data.min(),data.max(),100)
        pdf_values = stats.norm.pdf(x, mu, sigma)
        axes[i,1].plot(x,pdf_values,label = criteria)
        axes[i,1].legend(fontsize=5)

    # Box plot (right column)
    df.boxplot(column="MonthlyRate", by=criteria, ax=axes[i, 2])
    axes[i, 2].set_title(f"MonthlyRate Distribution by {criteria}")
    axes[i,2].tick_params(axis='x', labelrotation=45)

fig.suptitle("Average MonthlyRate of each Criteria",fontsize=16)
# Adjust layout
plt.tight_layout()
plt.show()
```

Using both bar plot to illustrate the difference in average monthly rate by each criteria. Also box plot to see other statistics such as range, quantile and mean

we could result that there are differents in average monthlyRate of each criteria such as `gender`,`maritalStatus` and `EducationField`

but the range of the monthlyRate of each group are quite similar

## With Control Variable

```python
df_grouped = df.groupby(["Gender", "MaritalStatus","EducationField"]).agg(avg_rate=("MonthlyRate","mean")).reset_index()
df_grouped
```

```python
def plot_subcriteria(df,main_criteria,sub1,sub2):
    '''
        main -> row
        sub1 -> column
        sub2 -> side by side column
    '''

    width = 0.7 / df[sub2].unique().size
    nrows = df[main_criteria].unique().size
    fig, axes = plt.subplots(nrows, figsize=(12, 12))
    fig.tight_layout(pad=1.0)
    
    index = np.arange(df[sub1].unique().size)
    for i,item in enumerate(df[main_criteria].unique()):
        main_df = df.loc[df[main_criteria] == item]
        for j,subitem in enumerate(df[sub2].unique()):
            # display(main_df.loc[main_df[sub2]==subitem])
            axes[i].bar(index + width*j,main_df.loc[main_df[sub2]==subitem].avg_rate,width,label=subitem)
            axes[i].set_title(f"{main_criteria} : {item}")
            axes[i].set_ylabel("Average MonthlyRate")
            axes[i].legend(fontsize=5)
    axes[-1].set_xticks(index-width/2)
    axes[-1].set_xticklabels(main_df[sub1].unique())
    axes[-1].tick_params(axis='x',labelrotation=45)
```

```python
plot_subcriteria(df_grouped,"MaritalStatus","EducationField","Gender")
plot_subcriteria(df_grouped,"EducationField","Gender","MaritalStatus")
plot_subcriteria(df_grouped,"MaritalStatus","Gender","EducationField")
```
